# Training visualisation
In this notebook we look at training loss

```python
import pickle
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('epochs_losses.pkl', 'rb') as f:
    epochs_losses = pickle.load(f)
```

```python
def show_loss(loss_file, title):
    with open(loss_file, 'rb') as f:
        epochs_losses = pickle.load(f)
    
    epochs = len(epochs_losses)
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.violinplot(epochs_losses, showmeans=True)
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(epochs)])
    ax.set_xlabel('Epochs')
    ax.set_ylabel('Loss')
    ax.set_title(title)
```

```python
show_loss('epochs_losses_24.pkl', 'Losses at each epochs')
```
